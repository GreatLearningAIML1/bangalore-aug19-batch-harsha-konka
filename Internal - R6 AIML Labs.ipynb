# The Iris Dataset
The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

The dataset contains a set of 150 records under five attributes - petal length, petal width, sepal length, sepal width and species.

## Question 1

### Import dataset
- Import iris dataset
- Import the dataset using sklearn library

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

iris_df = load_iris()
iris_df
```

## Question 2

### Get features and label from the dataset in separate variable
- you can get the features using .data method
- you can get the features using .target method

```python
data = iris_df.data
target= iris_df.target
```

## Question 3

### Create train and test data
- use train_test_split to get train and test set
- set a random_state
- test_size: 0.25

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size= 0.25, random_state= 42)
```

## Question 4

### One-hot encode the labels
- convert class vectors (integers) to binary class matrix
- convert labels
- number of classes: 3
- we are doing this to use categorical_crossentropy as loss

```python
# OneHotEncoding of Train & Test Class Variables using Keras.
from keras.utils import to_categorical

y_train_binary = to_categorical(y_train, num_classes= 3,  dtype= 'int32')
y_train_binary
```

```python
y_test_binary = to_categorical(y_test, num_classes= 3, dtype= 'int32')
y_test_binary
```

## Question 5

### Initialize a sequential model
- Define a sequential model

```python
from tensorflow.python.keras.layers import Dense

from tensorflow.python.keras import Sequential

# Initialize the constructor
model = Sequential()
```

## Question 6

### Add a layer
- Use Dense Layer  with input shape of 4 (according to the feature set) and number of outputs set to 3
- Apply Softmax on Dense Layer outputs

```python
# Adding and input layer with 4 input shape

model.add(Dense(5, activation= 'softmax', input_shape=(4,)))

# Adding an output layer with 3 output shape with Softmax

model.add(Dense(3, activation='softmax'))
```

## Question 7

### Compile the model
- Use SGD as Optimizer
- Use categorical_crossentropy as loss function
- Use accuracy as metrics

```python
model.compile(loss='categorical_crossentropy',
              optimizer='SGD',
              metrics=['accuracy'])
```

## Question 8

### Summarize the model
- Check model layers
- Understand number of trainable parameters

```python
model.summary()
```

## Question 9

### Fit the model
- Give train data as training features and labels
- Epochs: 100
- Give validation data as testing features and labels

```python
history = model.fit(X_train, y_train_binary,epochs=100, batch_size=1, verbose=1)
```

## Question 10

### Make predictions
- Predict labels on one row

```python
y_pred = np.round(model.predict(X_test))
y_pred[0:10]
```

### Compare the prediction with actual label
- Print the same row as done in the previous step but of actual labels

```python
score = model.evaluate(X_test, y_test_binary,verbose=1)

print(score)
```

## Question 7

### Define weight and bias
- Initialize weight and bias with tf.zeros
- tf.zeros is an initializer that generates tensors initialized to 0

```python
weights = tf.zeros((4, 1), dtype=tf.dtypes.float32)
```

```python
bias = tf.zeros(shape= (1000, 1), dtype=tf.dtypes.float32)
```

## Question 8

### Get prediction
- Define a function to get prediction
- Approach: prediction = (X * W) + b; here is X is features

```python
def get_prediction(X, W, b):
  return tf.add(tf.multiply(X, W), b)
```

```python
learning_rate = 0.01
training_epochs = 100
```

```python
y_pred = tf.add(tf.matmul(X_norm_tensor, weights), bias) # Equivalent to get_prediction method.
```

```python
y_pred
```

### Calculate loss
- Calculate loss using predictions
- Define a function to calculate loss
- We are calculating mean squared error

```python
# Mean Squared Error Cost Function
cost = tf.losses.mean_squared_error(y_numpy, y_pred)
```

## Question 9

### Define a function to train the model
1.   Record all the mathematical steps to calculate Loss
2.   Calculate Gradients of Loss w.r.t weights and bias
3.   Update Weights and Bias based on gradients and learning rate to minimize loss

```python
# Gradient Descent Optimizer 
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate)
  
# Global Variables Initializer 
init = tf.compat.v1.global_variables_initializer()
```

## Question 10

### Train the model for 100 epochs 
- Observe the training loss at every iteration

```python
from tensorflow.python.keras.layers import Dense

from tensorflow.python.keras import Sequential

def build_model():
  model = Sequential([
    Dense(64, activation='relu', input_shape=[len(X_train_stock.keys())]),
    Dense(64, activation='relu'),
    Dense(1)
  ])

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
history = model.fit(X_train_stock, y_train_stock, epochs=100, batch_size=1, verbose=1)
```

```python
score = model.evaluate(X_test_stock, y_test_stock,verbose=1)

print(score)
```
