```python
import pandas as py
import numpy as np
```

```python
#1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
Laliga = py.read_csv("Laliga.csv")
Laliga.columns = Laliga.iloc[0]
Laliga = Laliga.reindex(Laliga.index.drop(0)).reset_index(drop=True)
Laliga.columns.name = None
Laliga.replace(to_replace=['-'],value=0,inplace=True)
Laliga.style.hide_index()
```

```python
#2. Print all the teams which have started playing between 1930-1980.
Laliga[Laliga['Debut'].str[:4].between('1930', '1980')]
```

```python
#3. Print the list of teams which came Top 5 in terms of points
Laliga['Points'].astype(int)
Laliga.sort_values('Points', ascending=False).head(5)['Team']
```

```python
#4. Write a function with name Goal_diff_count which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

def diff_goal(frame):
    frame['GoalDiff'] = frame["GoalsFor"].astype(int) - frame["GoalsAgainst"].astype(int)
    return (frame)

goal_df=diff_goal(Laliga).sort_values('GoalDiff', ascending=False)
goal_df[['Team','GoalDiff']].head(1)
goal_df[['Team','GoalDiff']].tail(1)
```

```python
#5. Create a new column with name Winning Percent and append it to the data set 

def winning_percent(frame):
    frame['winning_percent'] = (frame['GamesWon'].astype(float) / frame['GamesPlayed'].astype(float) * 100)
    frame_sorted=frame.sort_values('winning_percent', ascending=False).replace(np.nan,'0%')
    return frame_sorted

win_p=winning_percent(Laliga)
win_p['Team'].head(5)
```

```python
#6. Group teams based on their Best position and print the sum of their points for all positions
def total_points(frame, all_positions):
    frame['Total Points'] = frame[all_positions].astype(int).sum(axis=1)
    frame_sorted=frame.sort_values('Total Points', ascending=False)
    return frame_sorted

total_p=total_points(Laliga,['Champion','Runner-up','Third','Fourth','Fifth', 'Sixth'])
pos_g = total_p.groupby('BestPosition')['Total Points'].sum().reset_index(name ='Total Points')
t=pos_g.sort_values(by=['Total Points'], ascending=False, inplace=False)
t.style.hide_index()
```
